```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import KFold
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
%matplotlib notebook
input_file = "spambase.data"
df = pd.read_csv(input_file, header = None)

X = df.iloc[:,0:57]
Y = df.iloc[:,57]
print 'Feature data shape: ',X.shape, 'label data shape: ',Y.shape
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.2, random_state=0)
print 'First split the data into Train and Test...'
print 'Train data shape: ', train_X.shape, train_Y.shape, 'Test data shape: ',test_X.shape ,test_Y.shape
kfold = KFold(5, True, 1)
fold_i = 1
err_array_per_fold = np.zeros(5)
len_array_per_fold = np.zeros(5)
print 'Run 5-fold on Train data...', 'sample num: ',train_X.shape[0]
print '\n'
for train_idx, valid_idx in kfold.split(train_X):
    train_x = train_X.iloc[train_idx]
    valid_x = train_X.iloc[valid_idx]
    train_y = train_Y.iloc[train_idx]
    valid_y = train_Y.iloc[valid_idx]
    #print train_x.shape,valid_x.shape,train_y.shape,valid_y.shape
    lr = linear_model.LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial').fit(train_x,train_y)
    valid_y_est = lr.predict(valid_x)
    valid_y_len = float(len(valid_y))
    err_array_per_fold[fold_i-1] = sum(valid_y_est!=valid_y)
    len_array_per_fold[fold_i-1] = valid_y_len
    err = sum(valid_y_est!=valid_y)/valid_y_len
    fp = sum(((valid_y-valid_y_est)==-1)/valid_y_len)
    fn = sum(((valid_y-valid_y_est)== 1)/valid_y_len)
    print 'fold-%d ' % fold_i, 'error: ',err, 'FP: ',fp,'FN: ',fn 
    fold_i = fold_i + 1
print 'average error rates across all folds: ',sum(err_array_per_fold)/sum(len_array_per_fold)

print '\n'
print 'Run LR model on the whole Train dataset (size: %d) and test on Test dataset (size:%d)...' % (train_X.shape[0],test_X.shape[0])
lr_final = linear_model.LogisticRegression(random_state=0, solver='lbfgs',multi_class='multinomial').fit(train_X,train_Y)
print 'final error on test dataset:',sum(lr.predict(test_X)!=test_Y)/float(len(test_Y))
 
```
